**Variable**

Rules for variable name
* Can contain only a-z, A-Z, 0-9 and _
* Cannot start with a number
* Case sensitive

Basic types of *simple* parameters
* integer
* float
* string

```python
# Assign variables
# Display values and types of variables
number_1 = 5
number_2 = 5.0
name = "abcde"
print(number_1)
print(type(number_1))
print(type(number_2))
print(type(name))
```

```python
# Type of variable is automatically assigned
x = 1
y = 1.0
z = "a"
print(type(x))
print(type(y))
print(type(z))
```

```python
# Integer (whole number) and Float (real number)
i1 = 5
i2 = -3
print(i1)
print(i2)
print(type(i1))

f1 = 1.0
f2 = -35.9
f3 = 1.35e-7
print(f1)
print(f2)
print(f3)
print(type(f1))
```

```python
# Python integer *cannot* overflow
# Float can be too large (overflow) or too small (underflow)
largeint = 10**1000
print(largeint)

largefloat1 = 1.0e300
largefloat2 = 1.0e500
print(largefloat1)
print(largefloat2)

smallfloat1 = 1.0e-300
smallfloat2 = 1.0e-500
print(smallfloat1)
print(smallfloat2)
```

```python
# String is character data
x = "hello"
y = 'abc'
print(x)
print(type(x))
print(len(x))
```

```python
# Changing (casting) variable type
i1 = int(2.8)
i2 = int("3")
f1 = float(1)
f2 = float("3")
f3 = float("3.5")
s1 = str(2)
s2 = str(3.3)
print(i1)
print(i2)
print(f1)
print(f2)
print(f3)
print(s1)
print(s2)
```

```python
# Get information on a variable
x = 1
help(x)
```
