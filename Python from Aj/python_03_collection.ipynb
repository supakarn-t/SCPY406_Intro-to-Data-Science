**Collection**

Four main types of collections (data structures)
* List (ordered, indexed group of objects)
* Tuple (similar to list, not mutable)
* Dictionary (unordered, indexed, but with key instead of number)
* Set (unordered, unindexed, unique elements)

```python
# List, indexed and changeable
# created by square brackets
# members can be anything, different variable types, even another list
x = [1, 2, 3]
y = [1, 1.5, "a"]
z = [1, 1.5, "a", [1, 2]]
print(x)
print(y)
print(z)
print(type(x))
```

```python
# Access member by index
x = [1, 2, 3, 4, 5]
print(x[0]) # index starts from 0
print(x[1])
print(x[-1]) # negative index means from the last
print(x[-2])
```

```python
# Access a bunch of member by slicing
# a:b = from a up to b (but not include b)
x = [1, 2, 3, 4, 5]
print(x[0:3])
print(x[1:4])
print(x[2:-2])
print(x[2:]) # from 0 - last
print(x[:3]) # from 0 - 3)
```

```python
# Finding length of a list
x = [1, 1.5, "a", [1, 2]]
print(len(x))
print(len(x[3]))
```

```python
# Changing value of member
x = [1, 2, 3, 4]
print(x)
x[0] = 5
print(x)
x[-1] = 2
print(x)
x[2:4] = [1, 2]
print(x)
```

```python
# Add an item to the end
x = [1, 2, 3]
x.append(4)
print(x)
```

```python
# Insert an item into list
# first number is position, second is content
x = [1, 2, 3]
x.insert(1, 5)
print(x)
```

```python
# Remove an item
x = [1, 2, 3, 4]
x.remove(2) # remove by value
print(x)
x.pop(2) # pop by index
print(x)
```

```python
# Print all items in the list
x = [1, 2, 3]
for item in x: # looping through item
  print(item)
for i in range(len(x)): # looping through list index
  print(x[i])
```

```python
# Check if something is in a list
x = [1, 2, 3]
print(1 in x)
print(0 in x)
```

```python
# Find index of an item
x = [1, 2, 3, 4]
print(x.index(3))
```

```python
x = [1, 2]
help(x)
```

```python
# Reverse the list order
x = [1, 2, 3, 4]
x.reverse()
print(x)
```

```python
# Another way to reverse the list
x = [1, 2, 3, 4]
x = x[::-1]
print(x)
```

```python
# Sort the list content
x = [1, 3, 2, 5, 4]
x.sort()
print(x)
x.sort(reverse=True)
print(x)
```

```python
# Careful when making copy of a list
x = [1, 2, 3]
y = x
x[0] = 0
print(x)
print(y)
print(y is x)
```

```python
# A way to avoid the copy problem
x = [1, 2, 3]
y = x[:] # slice to make python knows
x[0] = 0
print(x)
print(y)
print(y is x)
```

```python
# Another way to avoid the copy problem
x = [1, 2, 3]
y = list(x) # make new list from old one
x[0] = 0
print(x)
print(y)
print(y is x)
```

```python
# Tuple, indexed but not changeable
# Created with round brackets
# Pretty much the same as list, but cannot be changed
x = (1, 2, 3)
print(x)
print(type(x))
print(x[0])
print(x[-1])
```

```python
# Dictionary, unordered, but indexed and changeable
# Similar to list, but use key instead of index
# defined with curly bracket
x = {"key1": 1,
    "key2": 2,
    "key3": "a"}
print(x["key1"])
print(x["key2"])
print(x["key3"])
print(type(x))
```

```python
# Set, unordered, unindexed
# Take non-repeating component
x = set([1, 1, 1, 2, 2, 0, "a", "b", "a"])
print(x)
print(type(x))
```
