**Opratators**

Types of operators
* Arithmatic operators (+, -, *, /, exponentiation, modulo)
* Assignment operators
* Comparison operators (equal to, greater than, less than)
* Logical operators (and, or, not)
* Identity operators (is, is not)
* Membership operators (in, not in)

```python
# Arithmatic operators, used for mathematics
x = 4
y = 2
print(x + y) # addition
print(x - y) # subtraction
print(x * y) # multiplication
print(x / y) # division
print(x ** y) # exponentiation
print(x % y) # modulo
```

```python
# Careful of integer division!
# Integer division (//) ignores remainder
# Normal division (/) is smart enough to give float output
x = 1
y = 2
print(x // y) # integer division
print(x / y) # normal division
print(float(x) / y)
print(x / float(y))
```

```python
# Assignment operator, used to assign new value to variable
x = 1
y = x
z = x + y
print(z)
```

```python
# Can modify value based on old value
# Short-hand notation +=, -=, *=, /=, **=, %=
x = 1
x = x + 5
print(x)
x = 1
x += 5
print(x)
x *= 2
print(x)
x /= 3
print(x)
```

```python
# Comparison operator, compare two values, gives True/False
# ==, !=, >, >=, <, <=
x = 1
y = 1
z = 3
print(x == y)
print(x == z)
print(x > y)
print(x >= y)
```

```python
# Logical operator, used to combine True/False (and, or, not)
x = 1
y = 1
z = 3
print((x == y) and (x == z))
print((x == y) or (x == z))
print(not (x == y))
```

```python
# Identity operator (is, is not) checks whether objects are "the same"
# "Same" = same memory location, not same value
x = [1, 2, 3]
y = [1, 2, 3]
z = x
print(x is y)
print(x is z)
x[0] = 10
print(x)
print(y)
print(z)
```

```python
# Membership operator (in, not in) tell if an object is in a collection
x = [1, 2, 3]
print(1 in x)
print(4 in x)
```
