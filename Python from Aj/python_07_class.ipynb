**Class**

Python supports class and object-oriented programming.

Class can have variables ("attributes") and functions ("methods").

```python
# Defining the class
class Student():
  name = "John"
  age = 23

# Create a class "instance"
x = Student()
print(x)
print(x.name)
print(x.age)
help(x)

# Change value of attribute of this instance
y = Student()
print(y.name)
y.name = "James"
print(y.name)
```

```python
# Set values of attributes at creation by using __init__ function
class Student():
  def __init__(self, input_name, input_age):
    self.name = input_name
    self.age = input_age

x = Student("Jane", 25)
print(x.name)
print(x.age)
```

```python
# class can run functions ("methods")
class Student():

  # __init__ function to set attributes at creation
  def __init__(self, input_name, input_age):
    self.name = input_name
    self.age = input_age

  # Function that can use internal attribute values
  def greeting(self):
    print("Hello. My name is " + self.name + ".")

  # Function that can use internal attribute values
  def calculate_birth_year(self):
    year_of_birth = 2022 - self.age
    return year_of_birth

x = Student("Jane", 25)
x.greeting()
yob = x.calculate_birth_year()
print(yob)
```
