**Function**

A function is a piece of code that runs when it is called, useful to reuse the code or group them into separate parts.

Define a function using "def", the code of a function is in an indented block.

Function can take input parameters (can set default values).

A *simple* variable (number, string) is "passed by value", while more complicated variable (collection) is "passed by reference".

```python
# Define and call a simple function

def easy_function():
  print("this function is called!")
  print("Hello")

easy_function()
```

```python
# A function that takes an input parameter (argument)

def add_one(x):
  y = x + 1
  return y

a = 11.5
b = add_one(a)
print(a)
print(b)
```

```python
# A function that has default values for some input parameters

def add_numbers(x, y=1):
  z = x + y
  return z

a = 5
b = 10
c = add_numbers(a, b)
d = add_numbers(a)
print(c)
print(d)
```

```python
# A function can call itself (recursive)

def factorial(n):
  if n > 0:
    result = n * factorial(n-1)
  if n == 0:
    result = 1
  return result

print(factorial(5))
```

```python
# numbers are passed by value, collections are passed by reference

def func(x, y): # x is number, y is list
  x = x + 5
  y[0] = y[0] + 5

a = 0
b = [0, 0]
func(a, b)
print(a)
print(b)
```
