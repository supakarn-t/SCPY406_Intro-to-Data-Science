**Loop**

Used to execute a part of code multiple times

Two types of loop
* For loop
* While loop

Two extra control statements (break and continue)

```python
# for loop repeats the code a fixed amount of times
# range is a function used to run the index
for i in range(10):
  print(i)
for i in range(5, 10, 2):
  print(i)
```

```python
# can loop through items in collection, or through index
x = [10, 11, 12, 13]
for item in x:
  print(item)
for i in range(len(x)):
  print(i, x[i])
```

```python
# break statement stops and exits the loop
for i in range(10):
  if i == 5:
    print("stop printing at 5")
    break
  print(i)
```

```python
# continue statement stops the current iteration, but continue the loop
for i in range(10):
  if i == 5:
    print("skip printing at 5")
    continue
  print(i)
```

```python
# else command is executed after the loop is finished
for i in range(10):
  print(i)
else:
  print("finished printing")
```

```python
# for loops can be nested
for i in range(4):
  print("i = ", i)
  for j in range(4):
    print("  j = ", j)
```

```python
# while loop repeats the code while the condition is true
i = 0
while i < 10:
  print(i)
  i = i + 1
```

```python
i = 0
while True:
  print(i)
  i = i+1
  if i == 10:
    break
```
