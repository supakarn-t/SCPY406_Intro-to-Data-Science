**Conditional**

Used to direct the program directions (if, elif, else)

"Block" of code is indicated by indentations (no parenthesis)

```python
# "if" block is executed if the condition is met
a = 3
if a > 0:
  print("positive")
  x = 0
  print(x)
if a < 0:
  print("negative")
if a == 0:
  print("zero")
```

```python
# Conditions can be combined by logical operators
a = 3
if a > 0 or a == 0:
  print("positive or zero")
if not a > 0 and a != 0:
  print("negative")
```

```python
# "elif" (else if) block is checked if previous condition is not met
a = 3
if a >= 5:
  print("greater than 5")
elif a >= 3:
  print("between 3 and 5")
elif a >= 1:
  print("between 1 and 3")
```

```python
# "else" block catches anything that did not meet previous if/elif
a = -1
if a > 0:
  print("positive")
elif a == 0:
  print("zero")
else:
  print("not positive not zero -> negative")
```

```python
# if/elif/else can be nested by more indentation
a = 1
b = 1
if a > 0:
  print("a is positive, how about b?")
  if b > 0:
    print("b is positive too")
else:
  print("a is zero or negative, don't care about b")
```

```python
# Simple commands can be written in one line
a = 3
if a == 3: print("three")
```
