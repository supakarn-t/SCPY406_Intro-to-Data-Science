```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def data_gen_one_group(name, m, c, noise):

  # Make array in x, pretty random but fixed at 0,5,10
  x1 = np.linspace(0, 10, 3, dtype=float)
  x2 = np.random.rand(9) * 5.0
  x3 = np.random.rand(9) * 5.0 + 5.0
  x = np.append(x1, x2)
  x = np.append(x, x3)
  x = np.sort(x)
  x = np.around(x, decimals=1)

  # Make array in y, linear with noise, rounded to hundreds
  y = m*x + c + noise*np.random.randn(len(x))
  y = np.around(y, decimals=-1)

  # Make name array
  label = [name] * len(x)

  return label, x, y
```

```python
# Random seed, so that the random numbers are reproduceable
np.random.seed(0)

# Generate data for three companies with different parameters
name_a, x_a, y_a = data_gen_one_group("A", 3500, 30000, 2000)
name_b, x_b, y_b = data_gen_one_group("B", 3200, 40000, 2000)
name_c, x_c, y_c = data_gen_one_group("C", 4000, 25000, 2000)

# Plot to show data
plt.plot(x_a, y_a, 'ob', label="Company A")
plt.plot(x_b, y_b, 'or', label="Company B")
plt.plot(x_c, y_c, 'og', label="Company C")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend(loc="upper left")
plt.show()

# Print out in table format
print("# Company, Experience, Salary")
fmtstr = "      {0:1s}  ,   {1:4.1f}    ,  {2:5.0f}"
for i in range(len(name_a)):
  print(fmtstr.format(name_a[i], x_a[i], y_a[i]))
for i in range(len(name_b)):
  print(fmtstr.format(name_b[i], x_b[i], y_b[i]))
for i in range(len(name_c)):
  print(fmtstr.format(name_c[i], x_c[i], y_c[i]))
```
